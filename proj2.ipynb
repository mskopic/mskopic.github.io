```python
#Part1, Problem 1

import pandas as pd
import numpy as np
import sqlite3

sqlite_file = 'lahman2014.sqlite'
conn = sqlite3.connect(sqlite_file)

salary_query = "SELECT yearID, teamID, sum(salary) as total_payroll FROM Salaries WHERE yearID > 1989 GROUP BY yearID, teamID"
q1 = "SELECT yearID, teamID, franchID, G, W, ((CAST(W AS float)/CAST(G AS float)) * 100) as win_percentage "
q2 = "FROM Teams WHERE yearID > 1989 GROUP BY yearID, teamID"
qqq = q1 + q2
team_query = qqq

salaries = pd.read_sql(salary_query, conn)
teams = pd.read_sql(team_query, conn)
table = pd.merge(salaries, teams, on=['teamID', 'yearID'])
table.head()

## Commentary
#I used pandas dataframes to merge the two tables on ‘yearID’ and ‘teamID’ because we need information 
#on win percentage and total_payroll for each team for each year between 1990 and 2014. There was missing 
#data in the ‘teams’ table based on the fact that it has 728 rows, whereas the ‘salaries’ table has 730 
#rows. Each table was grouped by ‘yearID’ and ‘teamID’ for the reasons mentioned above. By merging on 
#these two columns, I ensured that the missing data in ‘salaries’ would not leave missing data in the final 
#‘table’ table.
```

```python
#Part 2, Problem 2
import matplotlib.pyplot as plt

plt.figure(num=None, figsize=(10, 8));
ax = plt.subplot(111)

#X and Y lists with keys of the same team name
years = dict.fromkeys(table['teamID'].tolist(), list())
team_payrolls = dict.fromkeys(table['teamID'].tolist(), list())

num_teams = len(team_payrolls.keys())

for i in range(0, num_teams) :
    team = list(team_payrolls.keys())[i]
    t_years = list()
    t_moolah = list()
    #print(team)
    for j in range(0, len(table.index)) :
        #index 2 is the total_payroll value, 1 is the team ID
        row_team = table.at[j, 'teamID']
        if(row_team == team) :
            t_years.append(table.at[j, 'yearID'])
            t_moolah.append(table.at[j, 'total_payroll'])
            
    team_payrolls[team] = t_moolah
    years[team] = t_years
    ax.plot(years[team], team_payrolls[team], label= team)
    
plt.xlabel('Year')
plt.ylabel('Total Payroll')
plt.title('MLB Total Payroll vs Year')
ax.legend(bbox_to_anchor=(1.2, 1.05))

plt.show()

##Commentary
#Code:
#I looped through the dataframe, once per team, looking for payroll and yearID to create points for
#each team and their payroll for each year. Then, I plotted each teams list in the same plot to show
#how each teams payrolls change over time. I needed to adjust the size of the plot to clearly show each team.

#Plot Description:
#The plot shows each teams payrolls changing with time between 1990 and 2014. Teams are denoted by colors
#explained in the legend to the right of the plot.

#Question 1:
#Based on the results of the plot of all team payrolls from 1990 to 2014, it appears that the variation of the 
#data increases over time. In other words, teams had more similar payrolls in 1990 than in 2014, when the range 
#of payrolls varied more than it had in every year prior. This is what I will be showing in problem 3, by finding
#the standard deviation of payrolls for each year between 1990 and 2014. Another claim this plot seems to support 
#is that the mean payroll for professional baseball teams increases over the 25 year period.
```

```python
#Part 2, Problem 3

#Show that the standard deviation increases over time
#For each year, calculate mean. Then, use the standard deviation equation to determine the deviation for each year
fig = plt.figure(num=None, figsize=(10, 8));
ax2 = plt.subplot(111)

yearly_payrolls = {'1990': [], '1991': [], '1992': [], '1993': [], '1994': [],
                   '1995': [], '1996': [], '1997': [], '1998': [], '1999': [],
                   '2000': [], '2001': [], '2002': [], '2003': [], '2004': [], 
                   '2005': [], '2006': [], '2007': [], '2008': [], '2009': [],
                   '2010': [], '2011': [], '2012': [], '2013': [], '2014': []}
num_years = len(list(yearly_payrolls.keys()))

#calculate means
yearly_pay_means = [] #means of all teams for a single year
prev_year = table.at[0,'yearID'] #should be 1990
summm = 0
num_vals = 0
for j in range(0, len(table.index)) :
    if table.at[j,'yearID'] != prev_year or j == (len(table.index) - 1):
        if num_vals != 0 :
            yearly_pay_means.append(summm / num_vals)
        num_vals = 0
        summm = 0
        prev_year = table.at[j,'yearID']
    
    num_vals += 1
    summm += float(table.at[j, 'total_payroll'])

#calculate Standard Deviation
yearly_variance = []
prev_year = table.at[0,'yearID'] #should be 1990
var_sum = 0
num_vals = 0
index = 0
for j in range(0, len(table.index)) :
    if table.at[j,'yearID'] != prev_year or j == (len(table.index) - 1):
        if num_vals != 0 :
            yearly_variance.append(pow(var_sum / num_vals, 0.5))
        num_vals = 0
        var_sum = 0
        if j != (len(table.index) - 1) :
            index += 1
        prev_year = table.at[j,'yearID']
    
    num_vals += 1
    var_sum += (float(table.at[j, 'total_payroll']) - yearly_pay_means[index]) ** 2

plt.plot(list(yearly_payrolls.keys()), yearly_variance)
plt.xlabel('Year')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviation in MLB Total Payrolls vs Year')
fig.autofmt_xdate()
plt.show()

##Commentary
#Code:
#I calculated the standard deviation for each year by looping through the dataframe twice. The first time,
#I calculated the mean payroll for each year. On the next loop, I used the mean of each year to find the 
#variance in the data, which when square rooted, gives the standard deviation of that years payrolls.

#Plot Explanation:
#The plot shows that the standard deviation in the data increases from 1990 to 2014. This is important
#information when evaluating the ability for teams to pay for wins because the disparity between payrolls
#has been increasing, which could mean that certain teams have a much greater disadvantage.
```

```python
#Part 3 Problem 5
yearly_variance #is the array with the std. deviation of each 
yearly_pay_means #array of yearly means for all teams in a year

standardized_payroll = list()


for i in range(0, len(table.index)) :
    year_i = int(float(table.at[i, 'yearID']) - 1990)
    standardized_payroll.append((table.at[i, 'total_payroll'] - yearly_pay_means[year_i]) / yearly_variance[year_i])
table['standardized_pay'] = standardized_payroll
table.head()

##Commentary:
#Using the given equation, I added a column in the dataframe for standardized payroll. I used lists created
#in previous steps (yearly_variance, showing each years standard deviation, and yearly_pay_means, showing the
#average payroll of a team per year) to calculate the standardized payroll for each year, then added it to
#the dataframe.
```
